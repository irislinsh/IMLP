### 01 - 奇偶數辨別（Odd or even）
#### 當使用者輸入一個的數字，它能夠辨別奇偶，並輸出檢驗結果給使用者。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一個數字 <br>
輸出(Output):奇數或偶數

* 進階: 可以限制輸入一個介於一定範圍（例如 1 到 1000）的數字

````python
message = input()
try:
    number = int(message)
    if number < 1 or number > 1000:
        print("Please enter an integer in between 1 to 1000")
    elif number % 2 == 0:
        print("Even")
    elif number % 2 == 1:
        print("Odd")
except:
    print("Please enter an integer")
````

### 03 -  猜數字（Guess the number）

#### 若使用者猜錯，就詢問他們想繼續玩還是退出；若使用者猜對，就顯示祝賀訊息，

![](data\guess.png)


````python
import random

Min = 0
Max = 100
attempt = 0

n = random.randint(0, 100)
    
while True:

    guess = int(input("猜數字範圍 " + str(Min) + " < ? < " + str(Max) + " :"))

    if guess <= Min or guess >= Max:
        attempt = attempt + 1
        print("Please enter a valid number")
        
    elif guess == n:
        print("賓果! 猜對了，答案是 " + str(n))
        break
        
    elif guess > n:
        attempt = attempt + 1
        print("再小一點!! 您猜了 " + str(attempt) + " 次")
        Max = guess
        
    elif guess < n:
        attempt = attempt + 1
        print("再大一點!! 您猜了 " + str(attempt) + " 次")
        Min = guess
````

> 題目：<br>

### 06 - 及格名單（Pass list）
#### 請使用集合功能來完成各科級名單的判斷
```

英文及格的有：柯南、灰原、丸尾、野口、步美
以上已列出全班所有人
```

>請分別列出<br>
數學及格但英文不及格的同學名單<br>
數學不及格但英文及格的同學名單<br>
兩者皆及格名單<br>
* Hint:差集(減法)、交集


````python
mathS = {'柯南','灰原','步美','美環','光彦'}
englishS = {'柯南','灰原','丸尾','野口','步美'}

classB = mathS.union(englishS)

englishF = classB.difference(englishS)
mathF = classB.difference(mathS)

print("數學及格但英文不及格的同學名單:", mathS.intersection(englishF))
print("數學不及格但英文及格的同學名單:", mathF.intersection(englishS))
print("兩者皆及格名單:", mathS.intersection(englishS))


#a=m.intersection(e)
#b=m.difference(a)
#c=e.difference(a)
#a1=list(a)
#b1=list(b)
#c1=list(c)
#b1.sort()
#print(b1)
#c1.sort()
#print(c1)
#a1.sort()
#print(a1)
````

### 08 - 資料夾管理(shutil)
#### 說明 ####

```
在目前所在的目錄下建立一files資料夾

令使用者輸入一數字N，並在files資料夾中建立f1, f2… fN等N個資料夾後列出files的資料夾內容

將files資料夾裡的f1資料夾重新命名成folder1後再列出files的資料夾內容

移除files資料夾中的folder1後再列出files的資料夾內容

最後移除files資料夾
※須先退出files資料夾(os.chdir(../)) 
```


````python
import os
import shutil

#to avoid IOError: [Errno 2] No such file or directory
#os.chdir('/Users/<your username>/pytest/Unit01')

if os.path.exists('files'):
    shutil.rmtree('files')

os.mkdir('files')

n = int(input())
originalDir = os.getcwd()
parentDir = os.getcwd() + '/files'

for amount in range(n):
    directory = '/f' + str(amount + 1)
    path = parentDir + directory
    os.mkdir(path)

print("-------os.listdir-------")
for file in os.listdir(parentDir):
     print(file)

print("-------os.listdir-------")
os.chdir(parentDir)

for amount in range(n):
    oldN = 'f' + str(amount + 1)
    newN = 'folder' + str(amount + 1)
    os.rename(oldN,newN)

for file in os.listdir(parentDir):
     print(file)
        
        
#remove directories
os.chdir(originalDir)
shutil.rmtree('files')
````

### 09 - 回文判斷（Is a palindrome）
#### 請使用者輸入單字，判斷它是否為回文，也就是該單字前後對稱，例如 madam，從前讀到後或是從後讀到前的順序都是 madam。

> 題目：<br>
輸入(Input): 雨 滋 春 樹 碧 連 天  天 連 碧 樹 春 滋 雨 <br>
輸出(Output):The text you have entered is a palindrome!<br>
輸入(Input): 資工訓練班 <br>
輸出(Output):The text you have entered is not a palindrome.


````python

def convertInputString():
    #請使用者輸入一串字串，先轉換字串為統一小寫
    rawInput = input("\nPlease enter a word, phrase, or a sentence \nto check if it is a palindrome: ") 
    return rawInput.lower()

def stripAnalphabetics(string): 
    #去除相關標點符號
    analphabeticList = [" ", "-", ".", ",", ":", ";", "!", "?", "'", "\""] 
    for item in analphabeticList:
        string = string.replace(item, "")
    return string

def runPalindromeCheck(straightList):
    reversedList = straightList[::-1] 
    if straightList == reversedList: 
        print("The text you have entered is a palindrome!")
    else: 
        print("The text you have entered is not a palindrome.")

def main(): 
    print("\nPalindrome checker")
    message = convertInputString()
    message = stripAnalphabetics(message)
    runPalindromeCheck(message) 
    
main() 
````
